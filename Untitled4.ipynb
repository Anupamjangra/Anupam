```python
from googlesearch import search
data = search ('IITM',num_results=10)

title = 'data science road map'

data = search(title,num_results=20)

all_data = []

for i in data:
    all_data.append(i+'\n')

# Mode =x: it only create new file 

file = open(f'{title}.txt',mode = 'x')
file.writelines(all_data)
file.close()
print('done')
```

```python
from googlesearch import search
data = search ('IITM',num_results=10)

title = 'Chat gpt trending news and investment'

data = search(title,num_results=5)

all_data = []

for i in data:
    all_data.append(i+'\n')

# Mode =a: it will append new line

file = open(f'{title}.txt',mode = 'a')
file.write('=====================\n')
file.writelines(all_data)
file.close()
print('done')
```

```python
url = '''https://picsum.photos/1500/1080'''
import requests as r
n = 10
for i in range(n):
    img_data = r.get(url).content
    f = open(f'image_{i}.jpg','wb')
    f.write(img_data)
    f.close()
    from IPython import display as d
    display(d.Image(f'image_{i}.jpg',width=200,height=200))
```
